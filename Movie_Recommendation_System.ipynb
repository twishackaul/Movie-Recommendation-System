```python
import numpy as np
import pandas as pd
import difflib       #to get the closest match to the value.
from sklearn.feature_extraction.text import TfidfVectorizer   #converts textual data to numericals/meaningful vectors.
from sklearn.metrics.pairwise import cosine_similarity        #to find similarities for the user input.
```

Data Collection and Pre-Processing

```python
# loading data
movies = pd.read_csv("/content/movies.csv")
```

```python
# printing first 5
movies.head()
```

```python
# finding out no. of rows and columns
movies.shape
```

```python
# selecting the relavent features for recommendation

selected_features = ['genres', 'tagline', 'keywords', 'cast', 'director']
print(selected_features)
```

```python
# replacing null values with null strings
for features in selected_features:
  movies[features] = movies[features].fillna("")
```

```python
#combining the features

combined = movies["genres"]+" "+movies["tagline"]+" "+movies["keywords"]+" "+movies["cast"]+" "+movies["director"]
print(combined)
```

```python
# converting text data to feature vector

vectorizer = TfidfVectorizer()
feature_vector = TfidfVectorizer().fit_transform(combined)
print(feature_vector)
```

Cosine Similarity

```python
# getting the similarity confidence/score value using cosine similariy
similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the user input
movie_name = input("Enter your favourite movie: ")
```

```python
# creating a list of all the movie names
titles = movies["title"].tolist()
print(titles)
```

```python
# finding the best match for the user's movie input

best_match = difflib.get_close_matches(movie_name, titles)
print(best_match)
```

```python
close_match = best_match[0]
print(close_match)
```

```python
# finding the index of movie with title

index =   movies[movies.title == close_match]["index"].values[0]
print(index)
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index]))         #enumerate carries out loop in a list.
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting movies wrt similarity scores

sorting_sm = sorted(similarity_score, key = lambda x:x[1], reverse = True)        # x[0]=index; x[1]=ss; lambda x:x[1] = wrt ss.
print(sorting_sm)                                                                 # reverse = True -> descending order.
```

```python
# finding similar movies based on index

print("Movies suggested for you: ")

i = 1

for movie in sorting_sm:
  index = movie[0]
  index_title = movies[movies.index == index]['title'].values[0]       # to print the movie titles of the indices in the sorted list.
  if (i<30):
    print(i, ".", index_title)
    i = i+1
```

Movie Recommendation System

```python
# user input
movie_name = input("Enter your favourite movie: \n")

# list of all movie titles
titles = movies["title"].tolist()

# find close match
best_match = difflib.get_close_matches(movie_name, titles)
close_match = best_match[0]

# index of movie with title
index =   movies[movies.title == close_match]["index"].values[0]

# similarity score
similarity_score = list(enumerate(similarity[index]))

#sorting
sorting_sm = sorted(similarity_score, key = lambda x:x[1], reverse = True)

# finding similar movies based on index
print("Movies suggested for you: \n")

i = 1

for movie in sorting_sm:
  index = movie[0]
  index_title = movies[movies.index == index]['title'].values[0]         # to print the movie titles of the indices in the sorted list.
  if (i<30):
    print(i, ".", index_title)
    i = i+1
```
